```python
import xuleta2 as xu
import numpy as np
from __future__ import division
np.seterr(all='ignore')
#nodata 32767
```

```python
folder = "/media/denis/seagate/MODIS/R_MYD09Q1/h10v04/Surf_Ref_8Days_250m_v6/"
```

```python
list1 = xu.listfiles(folder+"b1_Red/",full=True)
list2 = xu.listfiles(folder+"b2_NIR/",full=True)
```

```python
for i,j in zip(list1,list2):
    b1,m1 = xu.TifToArray(i)
    b2,m2 = xu.TifToArray(j)
    b2 = np.where(b2==32767,np.nan,b2)
    b1 = np.where(b1==32767,np.nan,b1)
    ndvi = 10000*(b2-b1)/(b2+b1)
    ndvi = np.where(ndvi<1,np.nan,ndvi)
    xu.ArrayToTif(ndvi,i[-12:-4],folder+"ndvi/",Metadata=m1,Type=1)
xu.shemale()
```
